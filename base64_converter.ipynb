```
base64_text = ''
```

```
base64_text = base64_text.strip('data:image/png;base64,')
```

```
base64_text = 'i' + base64_text
image_dat = bytes(base64_text, encoding='utf-8')
image_dat
```

```
# image_dat = b''
import base64
with open("imageToSave.png", "wb") as fh:
    fh.write(base64.decodebytes(image_dat))
from IPython.display import Image
Image('imageToSave.png')
```
